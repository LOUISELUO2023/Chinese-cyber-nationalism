```python
import pandas as pd
import re
import jieba
from collections import Counter
```

```python
df = pd.read_csv(r'D:\博士\course sem3\Data Mining\nationalism\结果文件\日本数据\日本去重 1203抽样.csv', encoding='utf-8')
df.head()
```

```python
def clean_chinese_text(text):
    # 移除特殊字符和数字
    text = re.sub("(\\d|\\W)+", " ", text)

    # 分词
    seg_list = jieba.cut(text, cut_all=False)

    # 移除停用词
    clean_seg_list = [word for word in seg_list if word not in stop_words]

    return clean_seg_list

# 查找最常见的词语的函数
def get_most_frequent_words(segmented_texts, num_words=20):
    counter = Counter()
    for text in segmented_texts:
        counter.update(text)
    return counter.most_common(num_words)
```

```python
# 定义一个函数来读取停用词文件
def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        stopwords = set(file.read().splitlines())
    return stopwords

# 使用函数读取停用词文件
file_path = r'D:\博士\course sem3\Data Mining\nationalism\结果文件\日本数据\日本cn_stopwords.txt'  # 替换为你的停用词文件路径
stop_words = load_stopwords(file_path)

# 打印停用词集合，确认内容
#print(stop_words)

jieba.del_word('水排')
jieba.add_word('核污水')
jieba.add_word('核污染水')
```

```python
# 对 'Content' 列应用清洗和分词函数
df['微博正文'] = df['微博正文'].astype(str)
df['Segmented_Content'] = df['微博正文'].apply(clean_chinese_text)

# 获取最常见的词语
most_frequent_words = get_most_frequent_words(df['Segmented_Content'])

# 显示最常见的词语
print([word for word,freq in most_frequent_words])

# 将分词重新组合成句子
df['Segmented_Content'] = df['Segmented_Content'].apply(lambda x: ' '.join(x))
```

```python
#df['ip'].value_counts()
# 计算 IP 地址的出现次数
ip_counts = df['ip'].value_counts()

# 指定输出文件路径
output_file_path = r'D:\博士\course sem3\Data Mining\nationalism\结果文件\日本数据\日本ip计数.xlsx'

# 将计数结果保存为 Excel 文件
ip_counts.to_excel(output_file_path, engine='openpyxl')
```
